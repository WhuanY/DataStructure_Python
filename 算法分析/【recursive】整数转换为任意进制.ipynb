## 使用递归，整数转换为任意进制
- 问题分析：
    - 我们使用辗转相除法来将整数转换为任意进制
    - 基本条件是，当剩下的数小于进制数的时候，辗转相除法就结束了；
    - 把整数换成更小的数，条件使用是整除后的商进行新的辗转相除法。
    - 不断调用辗转相除法

```python
def tostr(n, base):
    convertString = "01234556789ABCDEF"
    if n >= base:
        return tostr(n//base, base) + convertString[n % base]
    else:
        return convertString[n]
```

```python
tostr(1234,16)
```
